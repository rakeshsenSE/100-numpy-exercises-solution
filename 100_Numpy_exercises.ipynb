#### 63. Create an array class that has a name attribute (★★☆)

````python
import numpy as np

class NameArray(np.ndarray):
  def __new__(cls,input_array,name="Named Array"):
    obj=np.asarray(input_array).view(cls)
    obj.name=name 
    return obj
  
  def __array_finalize__(self, obj):
    if obj is None: return

    self.name=getattr(obj,'name',"Named Array")

z=NameArray(np.arange(5), name="My_Vector")
z.name
z
````

#### 75. How to compute averages using a sliding window over an array? (★★★)

````python
import numpy as np
def window_aveg(arr,n):
  cumsum=np.cumsum(arr)
  cumsum[n:] = cumsum[n:] - cumsum[:-n]
  return cumsum[n -1:] / n

a=np.array([20,30,1,33,53,21])
n=3

result=window_aveg(a,n)

print(a)
print(result)
````

#### 78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)

````python
import numpy as np

p = np.array([1,5])
p0 = np.array([[0,0],[1,2],[2,3]])
p1 = np.array([[1,1],[4,2],[3,8]])

v = p1 - p0
w = p - p0

cross = np.abs(v[:,0] * w[:,1] - v[:,1] * v[:,0])
length = np.sqrt(np.sum(v**2 , axis=1))
distances = cross / length

print(distances)
````

#### 79. Consider 2 sets of points P0,P1 describing lines (2d) and a set of points P, how to compute distance from each point j (P[j]) to each line i (P0[i],P1[i])? (★★★)

````python
import numpy as np


P0 = np.array([[0, 0], [1, 2], [2, 3]])
P1 = np.array([[1, 1], [4, 6], [5, 8]])

P = np.array([[1, 5], [0, 2]])

v = P1 - P0


w = P[:, np.newaxis, :] - P0[np.newaxis, :, :] 


cross_product = np.abs(v[np.newaxis, :, 0] * w[:, :, 1] - v[np.newaxis, :, 1] * w[:, :, 0])

line_length = np.sqrt(np.sum(v**2, axis=1))


distances = cross_product / line_length[np.newaxis, :]

print(distances)
````

````python
import numpy as np

def extract_subpart( z , center , tar_shape , fill=0):
  result = np.full(tar_shape, fill_value=fill, dtype=z.dtype)
  r_offset = tar_shape[0] // 2
  c_offset = tar_shape[1] // 2
  z_r_start = max(0, center[0] - r_offset)
  z_r_end   = min(z.shape[0], center[0] + r_offset + 1)
  z_c_start = max(0, center[1] - c_offset)
  z_c_end   = min(z.shape[1], center[1] + c_offset + 1)
  res_r_start = z_r_start - (center[0] - r_offset)
  res_r_end   = res_r_start + (z_r_end - z_r_start)
  res_c_start = z_c_start - (center[1] - c_offset)
  res_c_end   = res_c_start + (z_c_end - z_c_start)
  result[res_r_start:res_r_end, res_c_start:res_c_end] = z[z_r_start:z_r_end, z_c_start:z_c_end]
  return result

z=np.arange(1,26).reshape(5,5)
cropped = extract_subpart(z, center=(0, 0), tar_shape=(3, 3), fill=0)

print(z)
print(cropped)

````

#### 85. Create a 2D array subclass such that Z[i,j] == Z[j,i] (★★★)

````python
import numpy as np

class SymmetricArray(np.ndarray):
    def __setitem__(self, index, value):
        if isinstance(index, tuple) and len(index) == 2:
            i, j = index
            super().__setitem__((i, j), value)
            super().__setitem__((j, i), value)
        else:
            super().__setitem__(index, value)

def symmetric_array(shape, dtype=float):
    return np.zeros(shape, dtype=dtype).view(SymmetricArray)

Z = symmetric_array((3, 3))

print( Z)
Z[0, 2] = 99
print(Z)
````
